#**Central Tendency of the Philly Park Size Data**

```python
#Calculating the mean and median
meanps = pparks['ACREAGE'].mean()
medps = pparks['ACREAGE'].median()
```

```python
#Printing the mean and median acreage of all Philly parks
print('The mean acreage in the data set is ', meanps)
print('The median acreage in the data set is ', medps)
```

**1. What can you conclude from the values of the mean and median you extracted?**

> The mean and median values of this data show that Philadelphia parks vary in acreage, but ultimately, the majority of parks have fewer acres and there are few outliers with significantly larger acreage. This is discernible from the large disparity between mean and median-- As the median portrays the value in the middle of the data set, we can assume that a majority of the values in the data set are clustered around the median value of 3.6 acres. However, because the mean is significantly larger than the median (at 19.95 acres), and the mean value takes into account every value in a given data set, we can assume that larger outliers are skewing the mean value to make it much larger than the median, thus showing that few Philadelphia parks have a large acreage.


**2. What can you say about the distribution of the ACREAGE variable in the data?**

> Based solely on the measures of central tendency of this data set, we can discern that the acreage variable in the data set is distributed broadly. Most parks are clustered around the smaller end of the scale, but there are still some outliers that are dispersed at much larger values. This is proven by the median and mean; because the median is the exact middlepoint of the data set, we can assume that many data points are clustered around it or below it. The mean is much larger than the median, which proves that there are large outliers in this data set skewing this measure. Ultimately, these measures of central tendency indicate that the majority of Philadelphia parks have fewer acres, but still few exist with a larger number of acres.

```python
#Importing packages for histogram
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Histogram of park size
ax = pparks['ACREAGE'].hist(bins=20)
ax.set_xlabel('Philly Park Size in Acres')
ax.set_ylabel('Number of Parks')
ax.set_title('Figure 1: Distribution of Park Sizes in Philadelphia, PA')
ax.axvline(meanps, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(medps, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([meanps, medps], labels = ['Mean', 'Median'])
```

**3. Describe what the graph means.**
> This graph represents the frequency and sizes of parks in Philadelphia. The x-axis represents size in acres, while the y-axis reprsents the frequency of parks-- Ultimately, the graph shows that the majority of Philadelphia parks are within 0 to 250 acres. More precisely, most parks are within 0 to 100 acres in Philadelphia. There are outliers in this data, as the graph shows that some parks exist within the 250-500 acres range, as well as beyond 1000 acres-- However, according to the y-axis, the frequency of parks at these sizes is very minimal. The graph also depicts two measures of central tendency, the mean and median. The median, depicted by the green line, is computed to be roughly 3.6 acres. The mean, depicted by the red line, is roughly 19.95 acres. Ultimately, the graph shows that the distribution of Philadelphia parks is positively skewed, or skewed to the right, as the median < mean, and the majority of Philly park sizes are clustered within one particular range on the graph.

#**Selecting a Subset of Parks**

```python
#Selecting parks smaller than 10 acres
ax = pparks[~(pparks["ACREAGE"]=="Unknown")]["ACREAGE"].unique()
smallparks = pparks.loc[pparks['ACREAGE'] < 10]
smallparks.head()
```

```python
#Mean and median of parks smaller than 10 acres
meansps = smallparks['ACREAGE'].mean()
medsps = smallparks['ACREAGE'].median()
print('The mean acreage for small parks in the data set is ', meansps)
print('The median acreage for small parks in the data set is ', medsps)
```

```python
#Histogram of parks smaller than 10 acres
ax = smallparks['ACREAGE'].hist(bins=30)
ax.set_xlabel('Philly Park Size in Acres')
ax.set_ylabel('Number of Parks')
ax.set_title('Figure 2: Distribution of Park Sizes Less Than 10 Acres in Philadelphia, PA')
ax.axvline(meansps, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(medsps, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([meansps, medsps], labels = ['Mean', 'Median'])
```

**6. What does this figure show?**
> This figure shows the distribution and frequency of parks in Philadelphia with a size of under 10 acres. The red line depicts the mean of this data set, at 2.9 acres, and the green line depicts the median at roughly 2.2 acres. Overall, this distribution is skewed to the right, as the median < mean, and the majority of parks within the 0 to 10 acre-range are clustered between 0 and 2 acres. This shows that, among parks less than 10 acres in Philadelphia, most parks have smaller acreages, though there are some outliers with larger acreages, thus skewing the mean upwards.

```python
# Grouping the Parks based on the parent name
group_park = pparks.groupby(['PARENT_NAME']).sum()
group_park
```

**7. *Using* sns.pairplot plot the pairplot of ACREAGE and Shape__Area variables. What is the relationship that you can see?**
> The sns.pairplot function generated a plot that examines the relationship between the variables of acreage and shape area. The generated plots show that these variables are positively correlated-- As acreage increases, area also increases. The inverse is also true: as area inreases, acreage also increases.

```python
#Pairing variables
pair = ['ACREAGE','Shape__Area']
pair
```

```python
#Plotting pairplot
sns.pairplot(group_park[pair], size=2.0)
```

```python
# Calculating mean and median of parks grouped by parent name
groupmean = group_park['ACREAGE'].mean()
groupmedian = group_park['ACREAGE'].median()
print('The mean of the grouped parks is:', groupmean)
print('The median of the grouped parks is:', groupmedian)
```

```python
#Visualizing the distribution of the parks grouped by parent name
ax = group_park['ACREAGE'].hist(bins=20)
ax.set_xlabel('Park Size in Acres')
ax.set_ylabel('Number of Parks')
ax.axvline(groupmean, color='red', linestyle='solid', linewidth=2, alpha= .8)
ax.axvline(groupmedian, color='green', linestyle='solid', linewidth=2, alpha= .8)
ax.legend([groupmean, groupmedian], labels = ['Mean', 'Median'])
pylab.title("Figure 3: Histogram showing the distribution of parks grouped by parent name")
```

**9. Explain what the graph means.**
> This graph represents the distribution of parks in Philadelphia as grouped and represented by the variable of parent name, rather than public name. The x-axis represents park size in acres, while the y-axis represents the frequency of parks in Philadelphia. According to the histogram, we can discern that there is a large frequency of parks in Philadelphia with acreages between 0 to 250 acres. Some outliers exist, as the histogram shows very minimal numbers between the 500-1000 acre range, and beyond. However, ultimately, the majority of parks in Philadelphia are clustered at the lower end of the scale. The mean and median of this data are depicted by the red and green line, and represent 26. 1 acres and 3.03 acres respectively. As the median < mean, and most parks are clustered to the left of the x-axis, we can tell that ultimately, Philadelphia parks, categorized by parent names, fall into a clustered, positively skewed distribution with minimal outliers beyond the 500 acre range.

**10. What could you measure about this new dataset obtained by grouping the data by PARENT_NAME that would allow you to investigate further the difference in the two dataset?**
> Gathering more measures of central tendency and dispersion could provide more insight in the differences between the grouped parks dataset and the original parks dataset. While we have been able to generate the means and medians of both, generating the standard deviations of these two datasets and comparing them, for example, could provide more insight on the distribution and frequency of acreage in the context of grouped vs. ungrouped parks.
```python
#Measuring the standard deviation of the group parks
groupstd = group_park['ACREAGE'].std()
print('The standard deviation of the grouped parks sizes is ', groupstd)
```

**11. Explain what this standard deviation value means.**
> Standard deviation is a statistical measure of dispersion which analyzes the spread of data relative to the mean of a data set. In this context, the standard deviation of the group park sizes is roughly 151.4, which indicates that data points are spread far from the mean, and thus, the distribution of this data is wide and dispersed. Data points falling within roughly 151.4 acres above and below the mean qualify as being one standard deviation away from the mean. The shape of the frequencies where data points fall within this given distribution indicate the neutral, positive, or negative skew of the data set. Given our knowledge of previously generated histograms, such as Figure 3, we can assume this distribution is positively skewed, as most grouped parks in Philadelphia are within 151.4 acres below the grouped park mean of 26.1 acres.
